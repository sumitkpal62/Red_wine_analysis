### Variance, standard variance and quartile

```python
var_values=df.var()
print(var_values)
```

```python
std_values=df.std()
print(std_values)
```

```python
Q_values=round(df.quantile([.25, .75]),2)
print(Q_values)
```

```python
Q_values.index=["Q1", "Q3"]
```

```python
Q_values
```

```python
IQR_values=Q_values.loc['Q3']-Q_values.loc['Q1']
print(IQR_values)
print(type(IQR_values))
```

```python
Q_values.loc['IQR']=IQR_values
```

```python
Q_values
```

> From the above we can find the Q1, Q3 and IQR values of features

### Coefficient of variation

```python
# Coefficient of variation is calculated by (standard variance/mean)*100

CV=(std_values/mean_values)*100
print(round(CV,2))
```

```python
pd.DataFrame([round(CV,2)], index=["CV"])
```

```python
skewness_df=pd.DataFrame([df.skew()], index=["Skewness"])
skewness_df
```

> From the above analysis we can see that all values are > 0, meant that all are right skewed distribution.

* Here we can also conclude that residual sugar and chlorides have most outliers in their records
* And density and pH are less skewed

### Kurtosis

```python
kurtosis_value=df.kurtosis()

kurtosis_df=pd.DataFrame([kurtosis_value], index=["Kurtosis"])

kurtosis_df
```

> Here we can observe that chloride and residual sugar feature have the highest peak on the otherhand, citric acid and alcohol have the lowest peak.

* Leptokurtic: chlorides and residual sugar
* Platykurtic: citric acid and alcohol

> Here we can cross verify the skewness and kurtosis of each feature through visualisation

```python
# Assuming your DataFrame 'df' is already loaded
# df = ...

# List of all the column names you want to plot
# NOTE: Ensure these column names exactly match those in your DataFrame
features = [
    'alcohol', 'chlorides', 'citric acid', 'density', 
    'fixed acidity', 'free sulfur dioxide', 'pH', 'quality', 
    'residual sugar', 'sulphates', 'total sulfur dioxide', 'volatile acidity'
]

# Set the figure size for the 4x3 grid
plt.figure(figsize=(24, 20))

# Define the grid layout (4 rows, 3 columns)
rows = 4
cols = 3

# Define the styling for the KDE line
kde_style = {'color': 'red', 'linewidth': 2.5}

# Loop through the features and plot them
for i, feature in enumerate(features):
    # i starts at 0, but subplot numbering starts at 1, so use i + 1
    plt.subplot(rows, cols, i + 1)
    
    # Use histplot to create the histogram and KDE
    sns.histplot(
        data=df, 
        x=feature, 
        kde=True,
        color='skyblue', # Histogram bar color
        line_kws=kde_style # Applies the red color to the KDE line
    )
    
    # Set the title, capitalizing the first letter for uniformity
    plt.title(feature.replace('_', ' ').title(), fontsize=14)
    plt.xlabel(feature.replace('_', ' ').title(), fontsize=12)
    plt.ylabel('Frequency', fontsize=12)

# Adjust spacing between plots for a clean look
plt.tight_layout()

# Display the entire grid of plots
plt.show()
```
